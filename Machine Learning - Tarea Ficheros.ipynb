## 1. Acceso a ficheros

- `File` es un tipo de objeto predefinido en Python (*built-in*).
- Permite acceder a ficheros desde programas Python.
- Los ficheros son de un tipo especial: son *built-in*, pero no son ni números, ni
  secuencias, ni *mappings*. Tampoco responden a operadores en expresiones.
- La función `open` permite crear objetos de tipo fichero.

Formato general para abrir un fichero:
```
afile = open(filename, mode)
```
- `mode` es opcional. Por defecto, los ficheros se abren en modo lectura.
- Los datos leídos de un fichero siempre se obtienen en formato *string*. Lo mismo ocurre
  con escritura.
- Los ficheros se deben cerrar invocando `close` (liberación de recursos).


````python
# Lectura desde fichero usando método 'read'. Devuelve todo el contenido del fichero.
mi_fichero = open('res/multiple_lines.txt')
print(mi_fichero.read())
mi_fichero.close()

# Explicación línea por línea:
# 1) open('res/multiple_lines.txt') abre el fichero en modo lectura (modo por defecto, 'r').
# 2) mi_fichero.read() lee TODO el contenido del fichero de una sola vez y lo devuelve
#    como un único string (incluyendo los saltos de línea '\n').
# 3) print(...) lo muestra en pantalla.
# 4) mi_fichero.close() libera el recurso del sistema operativo asociado al fichero abierto;
#    es importante hacerlo siempre para no dejar archivos "colgados".
# Resultado esperado:
# Primera línea del fichero
# Segunda línea del fichero
# Tercera línea del fichero
# (seguido de una línea en blanco extra, porque print añade su propio salto de línea
# además del que ya trae el texto leído)
````

````python
# Lectura línea a línea a través del bucle 'for'.
mi_fichero = open('res/multiple_lines.txt')
for linea in mi_fichero:
    print(f"{linea}")
mi_fichero.close()

# Explicación línea por línea:
# 1) Se abre el fichero igual que antes.
# 2) 'for linea in mi_fichero' aprovecha que un objeto fichero es ITERABLE: en cada
#    iteración devuelve la siguiente línea del archivo (incluyendo su '\n' final).
# 3) print(f"{linea}") imprime cada línea; como 'linea' ya trae su propio salto de línea
#    y print añade otro, aparece una línea en blanco extra entre cada línea impresa.
# 4) Se cierra el fichero al terminar.
# Resultado esperado:
# Primera línea del fichero
# (línea en blanco)
# Segunda línea del fichero
# (línea en blanco)
# Tercera línea del fichero
# (línea en blanco)
````

````python
# leer de un archivo a una lista.
mi_fichero = open('res/multiple_lines.txt')
lineas = mi_fichero.readlines()
mi_fichero.close()
print(lineas)

# Explicación línea por línea:
# 1) Se abre el fichero.
# 2) mi_fichero.readlines() lee TODO el contenido y lo devuelve como una LISTA de strings,
#    donde cada elemento es una línea del archivo (cada una conservando su '\n' final).
# 3) Se cierra el fichero.
# 4) print(lineas) muestra la lista completa, con las comillas y los '\n' visibles como
#    caracteres de escape.
# Resultado esperado:
# ['Primera línea del fichero\n', 'Segunda línea del fichero\n', 'Tercera línea del fichero\n']
````

- Resolución de paths independiente de plataforma con `os.path.join`.

````python
import os

ruta = os.path.join("res", "multiple_lines.txt")
print(ruta)

# Explicación línea por línea:
# 1) os.path.join("res", "multiple_lines.txt") construye una ruta de archivo combinando
#    las partes dadas, usando el separador correcto según el sistema operativo:
#    '\\' en Windows, '/' en Linux/Mac. Esto evita escribir rutas "a mano" que solo
#    funcionen en una plataforma.
# 2) print(ruta) muestra la ruta resultante.
# Resultado esperado (en Windows): res\multiple_lines.txt
# Resultado esperado (en Linux/Mac, como este entorno): res/multiple_lines.txt
````

````python
dataset_file_path = ["res", "multiple_lines.txt"]
ruta = os.path.join(*dataset_file_path)
print(ruta)

# Explicación línea por línea:
# 1) dataset_file_path es una lista con las partes de la ruta.
# 2) 'os.path.join(*dataset_file_path)' usa el operador de desempaquetado '*' para pasar
#    cada elemento de la lista como un argumento posicional separado a join(), equivalente
#    a escribir os.path.join("res", "multiple_lines.txt") directamente.
# 3) Es útil cuando la ruta se construye dinámicamente con un número variable de partes.
# Resultado esperado (en Linux/Mac): res/multiple_lines.txt
````

- Cierre automático con la sentencia `with`. Esta es la forma habitual de leer de un
  fichero en Python (*context manager*).

````python
with open(ruta) as mi_fichero:
    for linea in mi_fichero:
        print(linea, end='')

# Explicación línea por línea:
# 1) 'with open(ruta) as mi_fichero:' abre el fichero y garantiza que se cerrará
#    AUTOMÁTICAMENTE al salir del bloque 'with', incluso si ocurre un error dentro
#    (no hace falta llamar a close() manualmente).
# 2) for linea in mi_fichero recorre cada línea del archivo.
# 3) print(linea, end='') imprime cada línea SIN agregar un salto de línea adicional
#    (end='' reemplaza el salto de línea por defecto de print), ya que 'linea' ya trae
#    su propio '\n'. Así se evita la línea en blanco extra que aparecía en el ejemplo
#    anterior con el bucle for.
# Resultado esperado:
# Primera línea del fichero
# Segunda línea del fichero
# Tercera línea del fichero
````

- Se pueden abrir varios ficheros en un mismo `with`.

````python
# abrir varios ficheros en el mismo with
ruta1 = os.path.join("res", "one_line.txt")
ruta2 = os.path.join("res", "multiple_lines.txt")
with open(ruta1) as fichero1, open(ruta2) as fichero2:
    print(fichero1.readlines())
    print(fichero2.readlines())

# Explicación línea por línea:
# 1) ruta1 y ruta2 apuntan a dos archivos distintos.
# 2) 'with open(ruta1) as fichero1, open(ruta2) as fichero2:' abre AMBOS ficheros dentro
#    del mismo bloque with, separados por coma; los dos se cerrarán automáticamente al
#    salir del bloque.
# 3) fichero1.readlines() devuelve la lista de líneas de one_line.txt (un solo elemento).
# 4) fichero2.readlines() devuelve la lista de líneas de multiple_lines.txt (tres elementos).
# Resultado esperado:
# ['Esta es una única línea de texto.\n']
# ['Primera línea del fichero\n', 'Segunda línea del fichero\n', 'Tercera línea del fichero\n']
````

## 2. Modos de acceso

Al crear un objeto de tipo `File` se puede especificar el modo de acceso (lectura/escritura):

| Modo | Descripción |
|------|-------------|
| `r`  | Solo Lectura |
| `w`  | Solo Escritura (Borra si el archivo ya existe) |
| `x`  | Solo Escritura (Falla si el archivo ya existe) |
| `a`  | Crea Fichero (Si existe, lo abre y añade al final) |
| `r+` | Lectura y Escritura |
| `b`  | Se puede añadir a otros modos para acceso binario |
| `t`  | Acceso para archivos de texto (default) |


## 3. Acceso para escritura

````python
def crear_lista(tamanyo):
    lista = []
    for i in range(tamanyo):
        lista.append(str(i) + '\n')
    return lista

ruta = os.path.join("res", "a_dummy.txt")
with open(ruta, 'wt') as fichero:
    fichero.write('Cabecera del ciclo del for en el ejemplo\n')
    lista = crear_lista(10)
    fichero.writelines(lista)

print("archivo creado")

# Explicación línea por línea:
# 1) crear_lista(tamanyo) genera una lista de strings: "0\n", "1\n", ..., hasta tamanyo-1,
#    cada uno con su salto de línea incluido.
# 2) ruta apunta a un nuevo archivo 'res/a_dummy.txt'.
# 3) 'wt' significa modo escritura ('w') + texto ('t', que es el modo por defecto).
#    Si el archivo ya existía, su contenido anterior se BORRA por completo.
# 4) fichero.write(...) escribe una línea de cabecera manualmente.
# 5) lista = crear_lista(10) genera 10 líneas ("0" a "9").
# 6) fichero.writelines(lista) escribe TODAS las líneas de la lista de una vez (a diferencia
#    de write(), writelines() NO añade saltos de línea automáticamente; por eso cada
#    elemento de 'lista' ya trae su propio '\n').
# 7) Al salir del 'with', el archivo se cierra y guarda automáticamente.
# Resultado esperado: "archivo creado"
# (y el archivo res/a_dummy.txt queda con 11 líneas: la cabecera + los números 0 al 9)
````

### Buffering
- Por defecto, el texto que se transfiere desde el programa a un fichero no se guarda en
  disco inmediatamente: se almacena en un *buffer*.
- Acciones como cerrar un fichero o invocar el método `flush` fuerzan que el contenido del
  buffer se transfiera a disco.


````python
ruta = os.path.join("res", "FicheroParaEscritura.txt")
fichero_write = open(ruta, 'w')
fichero_write.write('foter')

fichero_read = open(ruta, "r")
print(fichero_read.readlines())

fichero_write.flush()
print(fichero_read.readlines())

fichero_write.close()
fichero_read.close()

# Explicación línea por línea:
# 1) Se abre 'FicheroParaEscritura.txt' en modo escritura ('w') y se escribe el texto 'foter',
#    pero SIN cerrar ni hacer flush todavía: el texto queda en el buffer de escritura,
#    aún no está garantizado que esté físicamente en disco.
# 2) Se abre el MISMO archivo en modo lectura, en un objeto file distinto (fichero_read).
# 3) El primer fichero_read.readlines() puede devolver una lista VACÍA (o incompleta),
#    porque el contenido escrito por fichero_write todavía podría no haberse volcado a disco.
# 4) fichero_write.flush() fuerza a que el buffer se escriba físicamente en disco de inmediato.
# 5) El segundo fichero_read.readlines() ya debería poder leer el contenido 'foter', aunque
#    en algunos sistemas también puede depender de la posición del cursor de lectura
#    (fichero_read ya había intentado leer una vez).
# 6) Se cierran ambos ficheros al final.
# Resultado esperado (puede variar ligeramente según el sistema operativo y el buffering):
# []
# ['foter']
# Nota: este ejemplo ilustra el concepto de buffering; el comportamiento exacto (si la
# primera lectura muestra [] o ya el contenido) depende de la implementación del sistema
# de archivos y del tamaño del buffer, por lo que no siempre es 100% determinista.
````

## 4. Archivos CSV
Python permite leer y escribir datos en formato CSV (*Comma-Separated Values*), un formato
popular en ciencia de datos.

````python
# tabla_operaciones.csv contiene valores separados por comas
import csv

ruta = os.path.join("res", "tabla_operaciones.csv")
with open(ruta, encoding="utf-8") as fichero:
    data_reader = csv.reader(fichero, delimiter=',')
    for linea in data_reader:
        print(linea[0] + ' ---- ' + linea[1])

# Explicación línea por línea:
# 1) import csv carga el módulo estándar para trabajar con archivos CSV.
# 2) Se abre el archivo CSV con codificación UTF-8 explícita (para mostrar correctamente
#    tildes como en "división").
# 3) csv.reader(fichero, delimiter=',') crea un objeto lector que interpreta cada línea
#    del archivo separando sus valores por comas, devolviendo cada fila como una LISTA
#    de strings (una lista por línea).
# 4) for linea in data_reader recorre cada fila leída.
# 5) linea[0] y linea[1] acceden a la primera y segunda columna de cada fila
#    (Operacion y Descripcion), y se concatenan con ' ---- ' entre ambas.
# Resultado esperado:
# Operacion ---- Descripcion
# a + b ---- suma a y b
# a - b ---- resta a menos b
# a / b ---- a dividido
# a // b ---- a dividido entre b (quitando decimales)
# a % b ---- devuelve el resto de la division a/b (modulus)
# a * b ---- a multiplicado por b
# a ** b ---- a elevado a b
# Nota: en el PDF original aparecía "divisiÃ³n" en vez de "división"; eso ocurría porque
# el archivo original se guardó en una codificación y se leyó/imprimió asumiendo otra
# distinta (un problema clásico de encoding). Aquí, al crear y leer el archivo consistentemente
# en UTF-8, el texto se muestra correctamente.
````

````python
ruta = os.path.join("res", "tabla_operaciones.csv")
f = open(ruta, encoding="utf-8")
data_reader = csv.reader(f, delimiter=',')

ruta_o = os.path.join("res", "4_tabla_operaciones.csv")
with open(ruta_o, 'w', encoding="utf-8", newline='') as f2:
    csv_writer = csv.writer(f2, delimiter='|')
    for line in data_reader:
        print(line)
        csv_writer.writerow(line)

f.close()

# Explicación línea por línea:
# 1) Se abre el CSV original para lectura y se crea un csv.reader sobre él (delimitador ',').
# 2) ruta_o apunta a un nuevo archivo de salida, '4_tabla_operaciones.csv'.
# 3) Se abre ese archivo de salida en modo escritura con 'newline=""' (buena práctica al
#    escribir CSV, para evitar líneas en blanco extra en Windows).
# 4) csv.writer(f2, delimiter='|') crea un escritor CSV que usará '|' como separador
#    (en vez de la coma original).
# 5) for line in data_reader recorre cada fila leída del CSV original (cada 'line' es
#    una lista, ej: ['Operacion', 'Descripcion']).
# 6) print(line) muestra esa fila tal cual se leyó.
# 7) csv_writer.writerow(line) escribe esa misma fila en el nuevo archivo, pero ahora
#    separada por '|' en lugar de ','.
# 8) f.close() cierra el archivo de lectura (el de escritura se cierra automáticamente
#    al salir del 'with').
# Resultado esperado (impreso en pantalla, una lista por fila):
# ['Operacion', 'Descripcion']
# ['a + b', 'suma a y b']
# ['a - b', 'resta a menos b']
# ['a / b', 'a dividido']
# ['a // b', 'a dividido entre b (quitando decimales)']
# ['a % b', 'devuelve el resto de la division a/b (modulus)']
# ['a * b', 'a multiplicado por b']
# ['a ** b', 'a elevado a b']
# Además, queda creado el archivo 'res/4_tabla_operaciones.csv' con el mismo contenido
# pero separado por '|' en vez de ','.
````

## 5. Docstrings
- Python permite adjuntar documentación a los objetos e inspeccionarla a través de la línea
  de comandos o durante la ejecución del programa.
- Los docstrings se almacenan en el atributo `__doc__` de cada objeto.
- El valor de dicho atributo se puede mostrar por medio de la función `help`.


````python
help(open)

# Explicación línea por línea:
# 1) help(open) muestra la documentación integrada de la función built-in 'open': su firma
#    completa (parámetros como file, mode, buffering, encoding, etc.) y una descripción
#    de su comportamiento.
# Resultado esperado: un bloque de texto de ayuda generado automáticamente por Python,
# describiendo la firma y el propósito de la función open(). El contenido exacto depende
# de la versión de Python instalada.
````

- No es necesario editar el atributo `__doc__` directamente. Para asociar un docstring
  a un objeto basta con escribir el texto (entre triples comillas) al principio de módulos,
  funciones o clases, antes del código ejecutable.

````python
def funcion_de_prueba():
    """Esta es la documentación de la función de prueba"""
    pass

help(funcion_de_prueba)

# Explicación línea por línea:
# 1) El string entre triples comillas justo después de 'def funcion_de_prueba():' es su
#    DOCSTRING: Python lo detecta automáticamente y lo guarda en el atributo __doc__ de
#    la función.
# 2) 'pass' es el cuerpo de la función (no hace nada).
# 3) help(funcion_de_prueba) muestra esa documentación junto con la firma de la función.
# Resultado esperado (aproximado):
# Help on function funcion_de_prueba in module __main__:
#
# funcion_de_prueba()
#     Esta es la documentación de la función de prueba
````

- Accediendo directamente al atributo `__doc__` se obtiene solo el docstring (sin formato extra).

````python
print(funcion_de_prueba.__doc__)

# Explicación línea por línea:
# 1) funcion_de_prueba.__doc__ accede directamente al atributo donde Python guardó el
#    docstring de la función definida en la celda anterior.
# 2) A diferencia de help(), esto NO agrega información adicional (firma, formato, etc.):
#    solo devuelve el texto tal cual se escribió en las triples comillas.
# Resultado esperado: "Esta es la documentación de la función de prueba"
````

Para entidades más grandes, `help` muestra el docstring dividido en secciones.

````python
import sys
help(sys)

# Explicación línea por línea:
# 1) import sys carga el módulo estándar 'sys'.
# 2) help(sys) muestra la documentación completa del módulo: su docstring general, y un
#    listado organizado en secciones (funciones, clases, atributos, etc.) definidos en él.
# Resultado esperado: un bloque extenso de documentación con varias secciones (NAME,
# MODULE REFERENCE, DESCRIPTION, FUNCTIONS, etc.); el contenido exacto varía según la
# versión de Python.
````

Los docstrings que ocupan más de una línea suelen tener estas partes:
- Resumen en la primera línea.
- Línea en blanco.
- Descripción detallada.
- Otra línea en blanco antes del código.

````python
def funcion_de_prueba2():
    """Esta es la línea para el resumen.

    Este es el párrafo donde se puede escribir una descripción más
    detallada de la función. Observa que el resumen y la descripción
    detallada van separados por una línea en blanco. También hay
    otra línea en blanco antes de empezar el código.
    """
    pass

help(funcion_de_prueba2)

# Explicación línea por línea:
# 1) El docstring de esta función sigue la convención de resumen + línea en blanco +
#    descripción detallada, que es una práctica recomendada para documentar funciones
#    de forma clara y consistente.
# 2) 'pass' es el cuerpo (vacío) de la función.
# 3) help(funcion_de_prueba2) muestra ese docstring completo, con su formato de párrafos.
# Resultado esperado (aproximado):
# Help on function funcion_de_prueba2 in module __main__:
#
# funcion_de_prueba2()
#     Esta es la línea para el resumen.
#
#     Este es el párrafo donde se puede escribir una descripción más
#     detallada de la función. Observa que el resumen y la descripción
#     detallada van separados por una línea en blanco. También hay
#     otra línea en blanco antes de empezar el código.
````

## 7. Soluciones propuestas

````python
def cadena_en_fichero(ruta, cadena):
    with open(ruta, encoding="utf-8") as fichero:
        contenido = fichero.read()
    return cadena in contenido

ruta_prueba = os.path.join("res", "multiple_lines.txt")
print(cadena_en_fichero(ruta_prueba, "Segunda"))
print(cadena_en_fichero(ruta_prueba, "Cuarta"))

# Explicación línea por línea (Ejercicio 1):
# 1) cadena_en_fichero(ruta, cadena) abre el fichero con un 'with' (se cierra automáticamente).
# 2) fichero.read() lee TODO el contenido en un único string.
# 3) 'cadena in contenido' usa el operador 'in' para comprobar si 'cadena' aparece como
#    substring dentro del contenido leído; retorna True o False.
# 4) Se prueba buscando "Segunda" (aparece en "Segunda línea del fichero") y "Cuarta"
#    (no aparece en ningún lado del archivo de ejemplo).
# Resultado esperado:
# True
# False
````

````python
def crear_fichero_con_n_elementos(lista, ruta_destino, n):
    if n < 0 or n > len(lista):
        print(f"Error: n={n} está fuera de rango (0 a {len(lista)})")
        return
    elementos = lista[:n]
    with open(ruta_destino, 'w', encoding="utf-8") as fichero:
        for elemento in elementos:
            fichero.write(str(elemento) + '\n')
    print(f"Fichero creado con {n} elementos en {ruta_destino}")

lista_ejemplo = ['manzana', 'pera', 'uva', 'kiwi', 'mango']
ruta_salida = os.path.join("res", "primeros_elementos.txt")
crear_fichero_con_n_elementos(lista_ejemplo, ruta_salida, 3)
crear_fichero_con_n_elementos(lista_ejemplo, ruta_salida, 10)

# Explicación línea por línea (Ejercicio 2):
# 1) La función valida primero que 'n' esté en un rango válido: no puede ser negativo ni
#    mayor que la longitud de la lista; si está fuera de rango, imprime un error y
#    termina (return) sin crear el fichero.
# 2) 'lista[:n]' toma los primeros n elementos de la lista mediante slicing.
# 3) Se abre el fichero destino en modo escritura ('w'), lo que crea el archivo (o lo
#    sobrescribe si ya existía).
# 4) Se escribe cada elemento en una línea distinta, convertido a string por seguridad.
# 5) Se confirma con un mensaje cuántos elementos se escribieron.
# 6) La primera llamada usa n=3 (válido, dentro del rango 0-5) -> crea el fichero con
#    'manzana', 'pera', 'uva'.
# 7) La segunda llamada usa n=10, que excede la longitud de la lista (5) -> se detecta como
#    fuera de rango y NO se crea/sobrescribe el fichero.
# Resultado esperado:
# Fichero creado con 3 elementos en res/primeros_elementos.txt
# Error: n=10 está fuera de rango (0 a 5)
````

````python
def frecuencia_palabras(ruta):
    with open(ruta, encoding="utf-8") as fichero:
        contenido = fichero.read()
    palabras = contenido.split()
    frecuencias = {}
    for palabra in palabras:
        frecuencias[palabra] = frecuencias.get(palabra, 0) + 1
    return frecuencias

ruta_frase = os.path.join("res", "frase_ejemplo.txt")
with open(ruta_frase, 'w', encoding="utf-8") as f:
    f.write("es mejor que venga que que no venga")

print(frecuencia_palabras(ruta_frase))

# Explicación línea por línea (Ejercicio 3):
# 1) Se abre el fichero y se lee todo su contenido en un string.
# 2) contenido.split() separa el string en una lista de palabras, usando espacios en
#    blanco como delimitador por defecto (también maneja saltos de línea y tabulaciones).
# 3) 'frecuencias' es un diccionario vacío que irá acumulando los conteos.
# 4) for palabra in palabras recorre cada palabra de la lista.
# 5) 'frecuencias.get(palabra, 0) + 1' obtiene el conteo actual de esa palabra (0 si es
#    la primera vez que aparece) y le suma 1; el resultado se guarda de vuelta en el
#    diccionario para esa clave.
# 6) Se crea primero un archivo de prueba con la frase del enunciado, y luego se invoca
#    la función sobre ese archivo.
# Resultado esperado: {'es': 1, 'mejor': 1, 'que': 3, 'venga': 2, 'no': 1}
````
