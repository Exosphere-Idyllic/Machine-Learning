- Datos unidimensionales (similar a NumPy).

```python
# uso de masks para seleccionar
fibonacci = pd.Series([0, 1, 1, 2, 3, 5, 8, 13, 21])
print(fibonacci)
mask = fibonacci > 10
print(mask)
print(fibonacci[mask])

dst = pd.Series([13,21])
print(dst)
dst.equals(fibonacci)

fb = fibonacci[mask]
fb.reset_index(drop=True, inplace=True)
print(fb)
dst.equals(fb)

# Línea 1: comentario descriptivo.
# Línea 2: crea una Serie con los primeros números de Fibonacci (índice automático 0..8).
# Línea 3: la imprime.
# Línea 4: crea una máscara booleana: True donde el valor es mayor a 10.
# Línea 5: imprime la máscara.
# Línea 6: usa la máscara para filtrar solo los valores de Fibonacci mayores a 10 (conservando
#          sus índices originales 7 y 8).
# Línea 7: crea otra Serie 'dst' con valores [13,21] e índice automático 0,1.
# Línea 8: la imprime.
# Línea 9: dst.equals(fibonacci) compara ambas Series completas (valores E ÍNDICES); al tener
#          tamaños e índices distintos, el resultado es False (esta línea no imprime nada por
#          sí sola en un script, pero en Jupyter se muestra si es la última línea de la celda;
#          aquí hay más código después, por lo que su valor se descarta).
# Línea 10: filtra de nuevo fibonacci con la máscara, guardando en 'fb' (índices 7 y 8, valores 13,21).
# Línea 11: reset_index(drop=True, inplace=True) reindexa 'fb' para que use índices 0,1
#          consecutivos, descartando (drop=True) los índices originales 7,8, y modifica 'fb' in place.
# Línea 12: imprime 'fb' ya con índice reiniciado.
# Línea 13: dst.equals(fb) ahora compara dos Series con los mismos valores [13,21] y el mismo
#          índice [0,1]; al ser la última expresión de la celda, Jupyter muestra su resultado.
# Resultado esperado: se imprime la Serie fibonacci completa, la máscara booleana, los valores
#          filtrados con índices 7 y 8, la Serie dst, y finalmente 'fb' con índice 0,1 y valores
#          13,21; el resultado final mostrado por Jupyter es True (dst y fb ya coinciden).
```

### 3.2 Operaciones entre series

```python
# suma de dos series
# suma de valores con el mismo índice (NaN si no aparece en ambas)
serie1 = pd.Series([10,20,30,40], index=range(4) )
serie2 = pd.Series([1,2,3], index=range(3) )
suma = serie1 + serie2
print(suma)

# Línea 1-2: comentarios descriptivos.
# Línea 3: crea serie1 con 4 valores e índice 0..3.
# Línea 4: crea serie2 con 3 valores e índice 0..2.
# Línea 5: al sumar dos Series, pandas ALINEA por índice: suma los valores que comparten el
#          mismo índice, y donde un índice solo existe en una de las dos Series el resultado es
#          NaN (aquí, el índice 3 solo está en serie1).
# Línea 6: imprime el resultado de la suma.
# Resultado esperado:
# 0    11.0
# 1    22.0
# 2    33.0
# 3     NaN
# dtype: float64
```

```python
# resta de series (similar a la suma)
print(serie1 - serie2)

# Línea 1: comentario descriptivo.
# Línea 2: igual que la suma, la resta se alinea por índice; el índice 3 (solo en serie1) da NaN.
# Resultado esperado:
# 0     9.0
# 1    18.0
# 2    27.0
# 3     NaN
# dtype: float64
```

```python
# operaciones de pre-filtrado
result = serie1 + serie2
result[pd.isnull(result)] = 0 # mask con isnull()
print(result)

# Línea 1: comentario descriptivo.
# Línea 2: repite la suma alineada por índice (con NaN en la posición 3).
# Línea 3: pd.isnull(result) genera una máscara True donde el valor es NaN; se usa esa máscara
#          para reemplazar esos NaN por 0 directamente en 'result' (modificación in place).
# Línea 4: imprime el resultado ya sin valores NaN.
# Resultado esperado:
# 0    11.0
# 1    22.0
# 2    33.0
# 3     0.0
# dtype: float64
```

#### 3.2.1 Diferencias entre Pandas Series y diccionario

- Diccionario: estructura que relaciona claves y valores de forma arbitraria.
- Series: estructura estricta de valores con una lista de índices asignados por posición.
- Series es más eficiente para ciertas operaciones que los diccionarios.
- En las Series los valores de entrada pueden ser listas o NumPy arrays.
- En Series los índices semánticos pueden ser enteros o caracteres, igual que los valores.
- Series podría entenderse como algo intermedio entre una lista y un diccionario Python, pero de una dimensión.

### 4.3 Indexación y slicing con DataFrames

```python
d1 = {'ciudad':'Valencia', 'temperatura':10, 'o2':1}
d2 = {'ciudad':'Barcelona', 'temperatura':8}
d3 = {'ciudad':'Valencia', 'temperatura':9}
d4 = {'ciudad':'Madrid', 'temperatura':10, 'humedad':80}
d5 = {'ciudad':'Sevilla', 'temperatura':15, 'humedad':50, 'co2':6}
d6 = {'ciudad':'Valencia', 'temperatura':10, 'humedad':90, 'co2':10}
ls_data = [d1, d2, d3, d4, d5, d6] # lista de diccionarios
df_data = pd.DataFrame(ls_data, index = list('abcdef'))
display(df_data)

# Línea 1-6: se definen 6 diccionarios individuales, cada uno representando una fila; nótese
#          que NO todos tienen las mismas claves (algunos tienen 'o2', otros 'humedad', otros
#          'co2'), simulando datos heterogéneos o incompletos.
# Línea 7: se agrupan todos los diccionarios en una lista (una de las formas de crear un
#          DataFrame: lista de diccionarios).
# Línea 8: pd.DataFrame a partir de una lista de diccionarios crea automáticamente una columna
#          por cada clave distinta encontrada en cualquiera de los diccionarios; donde una fila
#          no tenga esa clave, el valor será NaN. Se asignan índices 'a' a 'f'.
# Línea 9: muestra la tabla resultante.
# Resultado esperado: tabla de 6 filas x 5 columnas (ciudad, temperatura, o2, humedad, co2),
#          con NaN en las combinaciones donde el diccionario original no tenía ese dato.
```

```python
# Acceso a un valor concreto por indice posicional [row, col]
print(df_data.iloc[1,1])
# Acceso a todos los valores hasta un índice por enteros
display(df_data.iloc[:3,:4])
# Acceso a datos de manera explícita, indice semantico (se incluyen)
display(df_data.loc['a', 'temperatura'])
display(df_data.loc[:'c', :'o2'])
display(df_data.loc[:'c', 'temperatura':'o2'])
display(df_data.loc[:, ['ciudad','o2']])

# Línea 1: comentario descriptivo.
# Línea 2: .iloc[1,1] accede por posición entera: fila 1 ('b'), columna 1 (temperatura).
# Línea 3: comentario descriptivo.
# Línea 4: .iloc[:3,:4] selecciona las primeras 3 filas (posiciones 0,1,2) y las primeras 4
#          columnas (posiciones 0..3), excluyendo el límite superior como en slicing normal.
# Línea 5: comentario descriptivo.
# Línea 6: .loc['a','temperatura'] accede por etiquetas semánticas: fila 'a', columna 'temperatura'.
# Línea 7: .loc[:'c', :'o2'] selecciona filas desde el inicio hasta 'c' INCLUSIVE, y columnas
#          desde el inicio hasta 'o2' INCLUSIVE (el slicing por etiquetas incluye ambos extremos).
# Línea 8: .loc[:'c', 'temperatura':'o2'] selecciona filas hasta 'c' inclusive, y solo las
#          columnas entre 'temperatura' y 'o2' inclusive.
# Línea 9: .loc[:, ['ciudad','o2']] selecciona todas las filas pero solo las columnas 'ciudad' y
#          'o2' (lista explícita de columnas).
# Resultado esperado: 8.0 (temperatura de la fila 'b'); una tabla 3x4; el valor 10 (temperatura
#          de 'a'); tablas parciales según los rangos de filas/columnas descritos, y finalmente
#          una tabla con solo 'ciudad' y 'o2' para las 6 filas.
```

```python
# indexación con nombre de columna (por columnas)
print(df_data['ciudad']) # --> Series
display(df_data[['ciudad', 'o2']])

# Línea 1: comentario descriptivo.
# Línea 2: indexar el DataFrame con un solo nombre de columna entre corchetes devuelve esa
#          columna como una Series.
# Línea 3: indexar con una LISTA de nombres de columnas devuelve un DataFrame con solo esas
#          columnas (aunque sea una sola columna en la lista, seguiría siendo un DataFrame).
# Resultado esperado: la columna 'ciudad' impresa como Series, seguida de una tabla con las
#          columnas 'ciudad' y 'o2' para las 6 filas.
```

```python
# indexación con índice posicional (no permitido!). Esto busca columna.
df_data[0]

# Línea 1: comentario descriptivo advirtiendo del comportamiento.
# Línea 2: df_data[0] NO se interpreta como 'fila en posición 0'; pandas intenta buscar una
#          COLUMNA cuyo nombre sea 0, y como no existe ninguna columna llamada 0, esto lanza un
#          KeyError.
# Resultado esperado: KeyError: 0 (excepción, no una tabla).
```

```python
# indexar por posición con 'iloc'
print(df_data.iloc[0]) # --> Series de la primera fila (qué marca los índices)

# Línea 1: comentario descriptivo.
# Línea 2: .iloc[0] accede a la fila en la posición 0 (la primera), devolviéndola como una
#          Series donde el índice de esa Series pasa a ser el NOMBRE DE LAS COLUMNAS originales.
# Resultado esperado: una Series con 'ciudad'->'Valencia', 'temperatura'->10, 'o2'->1.0, y NaN
#          para 'humedad' y 'co2'.
```

```python
# indexar semántico con 'loc'
df_data.loc['a'] # --> Series de la fila con índice 'a'

# Línea 1: comentario descriptivo.
# Línea 2: .loc['a'] accede a la fila con etiqueta 'a' (equivale a la primera fila en este
#          caso), devolviéndola como Series.
# Resultado esperado: idéntico al resultado de iloc[0] en este caso, ya que 'a' es la primera fila.
```

```python
# indexar semántico con 'loc'
df_data.loc[:'b'] # --> DataFrame de la fila con índice 'a' y 'b'

# Línea 1: comentario descriptivo.
# Línea 2: .loc[:'b'] selecciona todas las filas desde el inicio hasta la etiqueta 'b'
#          INCLUSIVE, devolviendo un DataFrame (no una Series) porque son varias filas.
# Resultado esperado: tabla con las filas 'a' y 'b'.
```

```python
df_data.loc[:'b'].loc[:,["o2", "humedad"]] # slicing anidado

# Línea 1: primero se seleccionan las filas 'a' y 'b' con .loc[:'b'], y sobre ese resultado
#          intermedio se aplica un segundo .loc para quedarse solo con las columnas 'o2' y
#          'humedad' (slicing/indexación encadenada).
# Resultado esperado: tabla de 2 filas ('a','b') x 2 columnas ('o2','humedad').
```

```python
# si se modifica una porcion del dataframe se modifica el dataframe original (referencia)
display(df_data)
serie = df_data.loc['a']
print(serie)
serie.iloc[2] = 3000 # setting with copy warning!!!
display(df_data)

# copiar data frame
df_2 = df_data.loc['a'].copy()
df_2.iloc[2] = 3000
display(df_2)
display(df_data)

# Línea 1: comentario descriptivo.
# Línea 2: muestra el DataFrame original completo.
# Línea 3: df_data.loc['a'] devuelve una Series que, según los casos, puede ser una VISTA sobre
#          los datos originales.
# Línea 4: imprime esa Series.
# Línea 5: modifica el tercer elemento (posición 2, columna 'o2') de la Series 'serie'; pandas
#          suele emitir un SettingWithCopyWarning porque no siempre es seguro saber si se está
#          modificando una vista o una copia.
# Línea 6: muestra df_data de nuevo: en muchos casos esta modificación SÍ se refleja en el
#          DataFrame original (comportamiento de referencia, no garantizado y por eso el warning).
# Línea 7: comentario descriptivo sobre cómo evitar el problema.
# Línea 8: se crea df_2 usando .copy() explícitamente, garantizando independencia de memoria.
# Línea 9: se modifica el mismo elemento en la copia.
# Línea 10: muestra df_2 modificado.
# Línea 11: muestra df_data, que esta vez permanece SIN cambios porque df_2 es una copia real.
# Resultado esperado: se observa cómo la primera modificación puede alterar el DataFrame
#          original (con warning de por medio), mientras que al usar .copy() explícitamente el
#          DataFrame original queda protegido de cambios posteriores.
```

```python
# ambos aceptan 'axis' como argumento
# df_data.iloc(axis=1)[0] # --> todos los valores asignados a la primera columna 'ciudad'
df_data.loc(axis=1)['ciudad'] # --> equivalente frame['ciudad']

# Línea 1: comentario descriptivo.
# Línea 2: (comentada) forma alternativa de usar iloc especificando axis=1 para referirse a
#          columnas en vez de filas; se deja comentada como referencia.
# Línea 3: .loc(axis=1)['ciudad'] es equivalente a df_data['ciudad'], usando axis=1 para indicar
#          explícitamente que se está indexando por columna.
# Resultado esperado: la columna 'ciudad' como Series, igual que df_data['ciudad'].
```

### 4.5 Slicing

```python
# slice por filas
d_and_d_characters = {'Name' : ['bundenth','theorin','barlok'], 'Strength' : [10,12,19], 'Wisdom' : [20,13,6]}
character_data = pd.DataFrame(d_and_d_characters, index=['a','b','c'])
display(character_data)
display(character_data[:-1])
display(character_data[1:2])

# Línea 1: comentario descriptivo.
# Línea 2: define un diccionario con datos de 3 personajes.
# Línea 3: crea el DataFrame con índices 'a','b','c'.
# Línea 4: muestra la tabla completa.
# Línea 5: indexar un DataFrame directamente con un slice (sin loc/iloc) actúa sobre las FILAS
#          por posición; [:-1] toma todas las filas menos la última.
# Línea 6: [1:2] toma solo la fila en la posición 1 (la segunda fila, 'theorin').
# Resultado esperado: tabla completa de 3 filas, luego una tabla de 2 filas ('a','b'), y luego
#          una tabla de 1 fila ('b', theorin).
```

```python
# slicing para columnas
display(character_data[['Name','Wisdom']])

# Línea 1: comentario descriptivo.
# Línea 2: indexar con una lista de nombres de columna selecciona solo esas columnas para
#          todas las filas.
# Resultado esperado: tabla de 3 filas x 2 columnas (Name, Wisdom).
```

```python
#slicing con 'loc' e 'iloc'
display(character_data.iloc[1:])
display(character_data.loc[:'b','Name':'Strength'])

# Línea 1: comentario descriptivo.
# Línea 2: .iloc[1:] selecciona todas las filas desde la posición 1 en adelante (excluye la
#          primera fila, 'a').
# Línea 3: .loc[:'b', 'Name':'Strength'] selecciona las filas hasta 'b' inclusive, y las
#          columnas entre 'Name' y 'Strength' inclusive.
# Resultado esperado: tabla con las filas 'b' y 'c'; y tabla con filas 'a','b' y columnas
#          'Name','Strength'.
```

¿Cómo filtrar filas y columnas? Por ejemplo, para todos los personajes, obtener 'Name' y 'Strength'

```python
# usando 'loc' para hacer slicing
display(character_data.loc[:,'Name':'Strength'])

# Línea 1: comentario descriptivo.
# Línea 2: .loc[:, 'Name':'Strength'] selecciona TODAS las filas y solo las columnas entre
#          'Name' y 'Strength' inclusive.
# Resultado esperado: tabla de 3 filas x 2 columnas (Name, Strength).
```

```python
# usando 'loc' para buscar específicamente filas y columnas
display(character_data.loc[ ['a','c'], ['Name','Wisdom'] ])

# Línea 1: comentario descriptivo.
# Línea 2: pasar dos listas explícitas a .loc selecciona el producto cruzado de esas filas
#          ('a','c') y columnas ('Name','Wisdom').
# Resultado esperado: tabla de 2 filas ('a','c') x 2 columnas (Name, Wisdom).
```

```python
# lo mismo con 'iloc'?
display(character_data.iloc[[0,2],[0,2]])
display(character_data.iloc[[0,-1],[0,-1]])

# Línea 1: comentario descriptivo (pregunta retórica).
# Línea 2: .iloc[[0,2],[0,2]] selecciona por posición las filas 0 y 2, y las columnas 0 y 2
#          (equivalente posicional al ejemplo anterior con loc).
# Línea 3: .iloc[[0,-1],[0,-1]] usa índices negativos: fila 0 y última fila (-1), columna 0 y
#          última columna (-1), una forma compacta de tomar "primero y último".
# Resultado esperado: dos tablas de 2x2, con las filas/columnas 'Name' y 'Wisdom' para los
#          personajes en las posiciones 0 y 2 (bundenth y barlok) en ambos casos.
```

```python
# lista de los personajes con el atributo Strength > 11
display(character_data.loc[character_data['Strength'] > 11, ['Name', 'Strength']])

# Línea 1: comentario descriptivo.
# Línea 2: character_data['Strength'] > 11 genera una máscara booleana por fila; .loc la usa
#          para filtrar filas, y además se especifica una lista de columnas ('Name','Strength')
#          a mostrar, combinando filtrado de filas y selección de columnas en un solo paso.
# Resultado esperado: tabla con 'theorin' (Strength=12) y 'barlok' (Strength=19), mostrando
#          solo sus columnas Name y Strength.
```
