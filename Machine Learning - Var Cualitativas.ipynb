```python
# Creación aleatoria de datos
import numpy as np
import random
random.seed(42)

# Explicación línea por línea:
# Línea 1: Importa la librería NumPy con el alias 'np', usada para crear arreglos y funciones numéricas (permutaciones, unique, etc.).
# Línea 2: Importa el módulo estándar 'random' de Python, usado para generar valores/semillas pseudoaleatorias.
# Línea 3: Fija la semilla del generador de 'random' en 42. Garantiza reproducibilidad SOLO para funciones de random.random,
#          random.randint, etc. Nota: np.random.permutation() (celda siguiente) usa el generador global de NumPy, NO el de
#          'random', por lo que el orden mezclado puede variar entre ejecuciones aunque random.seed(42) esté fijado.
#
# Resultado esperado: ninguna salida visible (celda de configuración inicial).
```

```python
nivel_estudios = []
nivel_estudios.extend(['Secundaria']*110)
nivel_estudios.extend(['Bachillerato']*172)
nivel_estudios.extend(['Grado']*145)
nivel_estudios.extend(['Master']*52)
nivel_estudios.extend(['Doctorado']*21)

# Explicación línea por línea:
# Línea 1: Crea una lista vacía 'nivel_estudios' que almacenará la variable cualitativa (categórica) del estudio.
# Línea 2: Agrega 110 repeticiones de 'Secundaria' con .extend() (une los elementos de la lista ['Secundaria']*110).
# Línea 3: Agrega 172 repeticiones de 'Bachillerato'.
# Línea 4: Agrega 145 repeticiones de 'Grado'.
# Línea 5: Agrega 52 repeticiones de 'Master'.
# Línea 6: Agrega 21 repeticiones de 'Doctorado'.
# Total: 110+172+145+52+21 = 500 elementos, agrupados por categoría (aún sin mezclar).
#
# Resultado esperado: ninguna salida visible; 'nivel_estudios' queda como lista de 500 strings agrupados por categoría.
```

```python
nivel_estudios = np.random.permutation(nivel_estudios)

# Explicación línea por línea:
# Línea 1: np.random.permutation() recibe la lista 'nivel_estudios' y devuelve un nuevo ndarray con los mismos 500
#          elementos pero en orden aleatorio (mezclados). Se reasigna a 'nivel_estudios', que pasa de lista de Python
#          a un np.ndarray de tipo string ('<U12').
#
# Resultado esperado: ninguna salida visible; 'nivel_estudios' ahora es un ndarray de 500 strings en orden aleatorio.
```

```python
print(nivel_estudios)

# Explicación línea por línea:
# Línea 1: Imprime el ndarray completo 'nivel_estudios' en pantalla, mostrando las 500 categorías ya mezcladas.
#
# Resultado esperado: un arreglo de NumPy con 500 strings (categorías) en orden aleatorio, mostrado en formato
#          array(['...', '...', ...], dtype='<U12').
```

```python
# DIAGRAMA DE BARRAS
import matplotlib.pyplot as plt
import numpy as np

estudios, frecuencia = np.unique(nivel_estudios, return_counts=True)
print(estudios, frecuencia)

# Explicación línea por línea:
# Línea 1: Importa pyplot como 'plt' para graficar.
# Línea 2: Vuelve a importar numpy (redundante pero inofensivo, ya estaba importado en la celda 1).
# Línea 3: np.unique(..., return_counts=True) devuelve dos arreglos: 'estudios' con las categorías únicas ordenadas
#          alfabéticamente y 'frecuencia' con el número de veces que aparece cada una (frecuencia absoluta).
# Línea 4: Imprime ambos arreglos para verificar los conteos.
#
# Resultado esperado: ['Bachillerato' 'Doctorado' 'Grado' 'Master' 'Secundaria'] [172 21 145 52 110]
#          (las categorías salen en orden alfabético junto con su frecuencia absoluta correspondiente).
```

```python
plt.ylabel('Frecuencia absoluta')
plt.xlabel('Nivel de estudios')
plt.title('Diagrama de barras')
plt.bar(estudios, frecuencia) # https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.show()

# Explicación línea por línea:
# Línea 1: Define la etiqueta del eje Y como 'Frecuencia absoluta'.
# Línea 2: Define la etiqueta del eje X como 'Nivel de estudios'.
# Línea 3: Define el título del gráfico.
# Línea 4: plt.bar() dibuja un diagrama de barras usando 'estudios' como categorías (eje X) y 'frecuencia' como
#          altura de cada barra (eje Y).
# Línea 5: Muestra la figura en pantalla.
#
# Resultado esperado: un gráfico de barras con 5 barras (Bachillerato, Doctorado, Grado, Master, Secundaria),
#          donde 'Bachillerato' es la más alta (172) y 'Doctorado' la más baja (21).
```

```python
# DIAGRAMA DE SECTORES
colores = {'red', 'blue', 'orange', 'green', 'cyan'}
plt.pie(frecuencia, labels=estudios, colors=colores, autopct='%1.2f%%')
plt.title('Diagrama de sectores')
plt.show()

# Explicación línea por línea:
# Línea 1: Define un 'set' (conjunto) con 5 nombres de colores para usar en el gráfico de pastel. Nota: al ser un
#          set, no garantiza un orden fijo de asignación de colores a cada categoría (podría variar entre ejecuciones);
#          para un control exacto color-categoría sería más seguro usar una lista en vez de un set.
# Línea 2: plt.pie() dibuja el diagrama de sectores (pastel) usando 'frecuencia' como tamaño de cada porción,
#          'estudios' como etiquetas, 'colores' como paleta y autopct='%1.2f%%' para mostrar el porcentaje de
#          cada sector con 2 decimales.
# Línea 3: Define el título del gráfico.
# Línea 4: Muestra la figura en pantalla.
#
# Resultado esperado: un gráfico circular con 5 sectores: Bachillerato 34.40%, Grado 29.00%, Secundaria 22.00%,
#          Master 10.40%, Doctorado 4.20% (porcentajes = frecuencia/500*100).
```

## Tarea: implementar con pandas esta información

El objetivo de esta celda era originalmente `# Tarea implementar con pandas esta información`. Lo que ya se calculó
manualmente con `np.unique(..., return_counts=True)` (tabla de frecuencias absolutas de una variable cualitativa) tiene
una solución equivalente ya existente en pandas: **`Series.value_counts()`**, que resuelve el conteo de frecuencias en una
sola línea, sin necesidad de reimplementar la lógica.

```python
import pandas as pd

# Tarea implementar con pandas esta información
conteo = pd.Series(nivel_estudios).value_counts()
print(conteo)

# Explicación línea por línea:
# Línea 1: Importa pandas con el alias habitual 'pd'.
# Línea 2: Solución en una sola línea: pd.Series(nivel_estudios) convierte el ndarray en una Serie de pandas, y
#          .value_counts() cuenta automáticamente cuántas veces aparece cada categoría, devolviendo el resultado ya
#          ordenado de mayor a menor frecuencia (equivalente pandas de np.unique(..., return_counts=True), pero sin
#          tener que ordenar ni emparejar manualmente los dos arreglos).
# Línea 3: Imprime el resultado.
#
# Resultado esperado (el orden exacto de las categorías con frecuencia empatada puede variar, pero los conteos son fijos):
# Bachillerato    172
# Grado           145
# Secundaria      110
# Master           52
# Doctorado        21
# Name: count, dtype: int64
```
